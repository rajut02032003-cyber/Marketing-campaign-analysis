```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("cleaned_data.csv")

# Convert date column to datetime
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])

# Create Age column
current_year = 2026
df['Age'] = current_year - df['Year_Birth']

# Create Age Groups
def age_category(age):
    if age < 20:
        return "Teenage"
    elif age >= 60:
        return "Super Senior"
    else:
        return "Senior"

df['Age_Group'] = df['Age'].apply(age_category)

# Filter data for 2014
df_2014 = df[df['Dt_Customer'].dt.year == 2014]

# Count shopping by age group
age_counts = df_2014['Age_Group'].value_counts()

# Plot bar chart
plt.figure()
age_counts.plot(kind='bar')

plt.title("Shopping Count by Age Group (2014)")
plt.xlabel("Age Group")
plt.ylabel("Number of Customers")
plt.xticks(rotation=0)

plt.show()
```

```python
plt.figure()
age_counts.plot(kind='pie', autopct='%1.1f%%')

plt.title("Shopping Distribution by Age Group (2014)")
plt.ylabel("")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("cleaned_data.csv")

# Convert date column
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])

# Create Age column
current_year = 2026
df['Age'] = current_year - df['Year_Birth']

# Create Age Groups
def age_category(age):
    if age < 20:
        return "Teenage"
    elif age >= 60:
        return "Super Senior"
    else:
        return "Senior"

df['Age_Group'] = df['Age'].apply(age_category)

# Extract year
df['Year'] = df['Dt_Customer'].dt.year

# Filter for 2012, 2013, 2014
df_filtered = df[df['Year'].isin([2012, 2013, 2014])]

# Create pivot table (counts)
comparison = df_filtered.pivot_table(
    index='Year',
    columns='Age_Group',
    aggfunc='size',
    fill_value=0
)

print(comparison)  # optional: see values

# Plot grouped bar chart
comparison.plot(kind='bar')

plt.title("Shopping Comparison by Age Group (2012-2014)")
plt.xlabel("Year")
plt.ylabel("Number of Customers")
plt.xticks(rotation=0)

plt.show()
```
